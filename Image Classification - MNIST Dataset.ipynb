```python
from keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import numpy as np
new_xtrain =np.reshape(x_train,(60000,784))
new_xtest =np.reshape(x_test,(10000,784))
print(new_xtrain.shape)
print(new_xtest.shape)
```

#### NAIVE BAYES MODEL

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
model.fit(new_xtrain, y_train)
pred = model.predict(new_xtest)
acc_score = accuracy_score(y_test, pred)
cm = confusion_matrix(y_test, pred)
```

```python
print("\nACCURACY  :  \n",acc_score)
print("\n\nCONFUSION MATRIX  :  \n",cm)
```

#### DECISION TREE MODEL

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(random_state =0 )
tree_model.fit(new_xtrain, y_train)
```

```python
pred = tree_model.predict(new_xtest)
acc_score = accuracy_score(y_test, pred)
cm = confusion_matrix(y_test, pred)
```

```python
print("\nACCURACY  :  \n",acc_score)
print("\n\nCONFUSION MATRIX  :  \n",cm)
```

#### RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier(random_state =0 )
forest_model.fit(new_xtrain, y_train)
```

```python
pred =forest_model.predict(new_xtest)
acc_score = accuracy_score(y_test, pred)
cm = confusion_matrix(y_test, pred)
```

```python
print("\nACCURACY  :  \n",acc_score)
print("\n\nCONFUSION MATRIX  :  \n",cm)
```
