```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
```

```python
data = pd.read_csv('daily-total-female-births-CA.csv',header=0,index_col=0)
```

```python
data.head()
```

```python
plt.plot(data)
```

```python
values = data.values
```

```python
res = adfuller(values)
```

```python
print('Augmented Dickey_fuller Statistic : %f' % res[0])
print('p-value: %f' % res[1])
```

```python
print('critical values at different levels:')
for k,v in res[4].items():
    print('\t%s: %.3f' %(k,v))
```

```python
data1 = np.log(data)
```

```python
values1 = data1.values
```

```python
res = adfuller(values1)
```

```python
print('Augemented Dickey_fuller Statistic: %f' %res[0])
print('p-value: %f' %res[1])
```

```python
data2 = np.sqrt(data)
```

```python
values2 = data2.values
```

```python
res = adfuller(values2)
```

```python
print('Augemented Dickey_fuller Statistic: %f' %res[0])
print('p-value: %f' %res[1])
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
m1=ARIMA(data1,order=(2,1,0))
result1=m1.fit()
```

```python
data_log_dif = data1 - data1.shift()
```

```python
plt.plot(data1)
plt.show()
```

```python
plt.plot(data_log_dif)
plt.show()
```

```python
plt.figure()
plt.plot(data_log_dif)
plt.plot(result1.fittedvalues,color='black')
```

```python
m1=ARIMA(data1,order=(0,1,2))
result2=m1.fit()
```

```python
plt.figure()
plt.plot(data_log_dif)
plt.plot(result2.fittedvalues,color='black')
```

```python
m1=ARIMA(data1,order=(2,1,2))
result3=m1.fit()
```

```python
result3.summary()
```

```python
plt.figure()
plt.plot(data_log_dif)
plt.plot(result3.fittedvalues,color='black')
```
